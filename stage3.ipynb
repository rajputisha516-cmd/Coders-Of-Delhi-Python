# Finding People You May Know

```python
import json 

# load data through functions
def load_data(filename):
    with open(filename , "r") as f:
        return json.load(f)

# main functions
def find_people_you_may_know(user_id , data ):
    user_friends = {}

    # find the list of friends for each user 
    for user in data['users']:
        user_friends[user['id']] = user['friends']
    print(user_friends) 

    # if given user_id is not present in the data then it returns empty list 
    if user_id not in user_friends :
        return []

    # Extract direct friends of a user 
    direct_friends = user_friends[user_id]
    print(direct_friends)

    # find mutual friends and give suggestions
    suggestions = {}
    for friend in direct_friends :
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends :
                # Count mutual friends 
                suggestions[mutual] = suggestions.get(mutual , 0 ) + 1
           
    # sort suggestions based on their high score 
    sorted_suggestions = sorted(suggestions.items() , key = lambda x : x[1] , reverse = True )

    # return suggested mutual friends id 
    return [ user_id for user_id , mutual_count in sorted_suggestions ]
    
data = load_data("data1.json")
data
user_id = 1
recommendation = find_people_you_may_know(user_id , data)
print(recommendation)
```
